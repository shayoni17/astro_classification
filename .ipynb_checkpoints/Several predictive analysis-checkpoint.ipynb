```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from scipy import signal
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error as MSE
from sklearn import svm
```

```python
data = pd.read_csv('Skyserver_SQL2_27_2018 6_51_39 PM.csv')
```

```python
data['u-g'] = data['u'] - data['g']
sns.scatterplot(x='z', y='u-g', hue='class', data=data)
```

# KNN

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['u', 'g', 'r', 'i', 'z']], data['class'], test_size=0.3, random_state=21)
```

```python
k_range = range(1, 26)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))

print(scores)
```

```python
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
knn.fit(X_train, y_train)
```

```python
knn_predict = knn.predict(X_test)
```

```python
accuracy_score(y_test, knn_predict)
```

```python
cm_knn = confusion_matrix(y_test, knn_predict)
```

```python
print(cm_knn)
```

# RF

```python
SEED = 1
```

```python
rf_range = range(300, 600, 25)
scores = []
for i in rf_range:
    rf = RandomForestClassifier(n_estimators = i, random_state = SEED)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))

print(scores)
```

```python
plt.plot(rf_range, scores)
plt.xlabel('Value of n_estimators for RF')
plt.ylabel('Testing Accuracy')
```

```python
rf = RandomForestClassifier(n_estimators = 450, random_state = SEED)
```

```python
rf.fit(X_train, y_train)
```

```python
rf_predict = rf.predict(X_test)
```

```python
accuracy_score(y_test, rf_predict)
```
